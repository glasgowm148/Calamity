###1. Read in the Crowdflower data set and extract the same keywords used for search

```python
import pandas as pd
```

```python
df = pd.read_csv("data/tweets/disaster-tweets.csv")
print df.columns
```

**Check out the keywords**

```python
keywords = df["keyword"].unique()
print keywords
```

**Clean up the keywords by replacing %20 with a space**

```python
import re
```

```python
keywords = keywords[1:]
```

```python
cleaned_keywords = [re.sub('%20', ' ', keyword) for keyword in keywords]
```
